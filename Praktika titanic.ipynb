```r
library(tidyverse)
list.files(path = "../input")
```

```r
library(Hmisc)
library(class)
library(caret)
library(corrplot)
```

```r
train <- read.csv("/kaggle/input/titanic/train.csv")
summary(train)
```

```r
hist(train$Age, 40)
```

```r
#Fixing NAs
train$Age <- impute(train$Age, "random")
```

```r
test <- read.csv("/kaggle/input/titanic/test.csv")
summary(test)
```

```r
hist(test$Age, 40)
```

```r
#Fixing NAs
test$Age <- impute(test$Age, "random")
test$Fare <- impute(test$Fare, median)
```

```r
#Converting factors
train$Sex <- as.numeric(train$Sex)
test$Sex <- as.numeric(test$Sex)
train$Age <- as.numeric(train$Age)
train$Embarked <- as.numeric(train$Embarked)
test$Embarked <- as.numeric(test$Embarked)
train$Ticket <- as.numeric(train$Ticket)
test$Ticket <- as.numeric(test$Ticket)
train$Cabin <- as.numeric(train$Cabin)
test$Cabin <- as.numeric(test$Cabin)
```

```r
train2 <- train[,c(2,3,5,6,7,8,9,10,11,12)]
test2 <- test[,c(2,4,5,6,7,8,9,10,11)]
head(train2,3)
head(test2,3)
```

```r
#Variables normalization
train2[-1] = scale(train2[-1])
test2 = scale(test2)
head(train2,3)
head(test2,3)
```

```r
train2_cor <- cor(train2)
train2_cor
```

```r
corrplot(train2_cor)
```

```r
# Encoding the target feature as factor
train2$Survived = factor(train2$Survived, levels = c(0, 1))
```

```r
train3 <- train2[,c(1,2,3,7,8,9,10)]
test3 <- test2[,c(1,2,6,7,8,9)]
```

```r
knn <- knn(train=train3[,-1], test=train3[,-1], cl=train3[,1], k=2)
```

```r
cm <- confusionMatrix(knn, train3[, 1])
cm
```

```r
Survived <- knn(train=train3[,-1], test=test3, cl=train3[,1], k=5)
```

```r
PassengerId <- test$PassengerId
to_csv <- data.frame(PassengerId, Survived)
write_csv(to_csv, 'r_k_nearest_neighbors.csv')
```
